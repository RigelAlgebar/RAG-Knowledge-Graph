# Data Cleaning

```python
queries_df['query'] = queries_df['query'].str.lower()
corpus_df['body'] = corpus_df['body'].str.lower()
```

```python
corpus_df['author'] = corpus_df['author'].fillna('Not Found')
```

```python
# Standardize the 'published_at' column
corpus_df['published_at'] = pd.to_datetime(corpus_df['published_at'], errors='coerce')
```

```python
# Drop rows with invalid dates
corpus_df = corpus_df.dropna(subset=['published_at'])
```

```python
# Remove duplicates in both datasets
queries_df = queries_df.drop_duplicates()
corpus_df = corpus_df.drop_duplicates()
```
